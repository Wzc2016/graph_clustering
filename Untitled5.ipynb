```python
import numpy as np
import time
from sklearn.cluster import SpectralClustering
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import fowlkes_mallows_score
from sklearn.metrics.cluster import adjusted_rand_score
import scipy.sparse as sp
from sklearn import mixture
import networkx as nx
import matplotlib.pyplot as plt
```

```python
edge = []
with open('email-Eu-core.txt', 'r') as infile:
    for i in infile.readlines():
        edge.append(list(map(int, i[:-1].split(' '))))
        
label = []
with open('email-Eu-core-department-labels.txt', 'r') as infile:
    for i in infile.readlines():
        label.append(int(i[-2]))
label = np.array(label)
```

```python
A = np.zeros((len(label), len(label)))

for pair in edge:
    A[pair[0], pair[1]] = 1
    A[pair[1], pair[0]] = 1
```

```python
N = 3

index = np.where(label < 3)[0]
label_ = label[index]
A_ = A[index]
A_ = A_[:, index]

G=nx.from_numpy_matrix(A_)
    
np.random.seed(122)
pos = np.random.randn(len(label), 10)

def reduce(pos):
    model = PCA(n_components=2)
    W = model.fit_transform(pos)
    H = model.components_
    return W

pos2 = reduce(pos)
diction = {}
for i in range(pos2.shape[0]):
    diction[i] = pos2[i]
    
color = ['#9e0142',
'#d53e4f',
'#f46d43',
'#fdae61',
'#fee08b',
'#e6f598',
'#abdda4',
'#66c2a5',
'#3288bd',
'#5e4fa2']

nodeColor = []
for i in label_:
    nodeColor.append(color[i * 3 - 1])
    
plt.figure(figsize=(20, 20))
nx.draw(G, diction,
        node_color=nodeColor,
        alpha = 0.8,
        with_labels=False,
        node_size = 50 )
```

```python
#Spectral Clustering

# norm_A = normalize(A)

cluster = SpectralClustering(n_clusters=3, affinity='precomputed').fit(A_)
print(fowlkes_mallows_score(label_, cluster.labels_))
```

```python
# ours

def normalize(mx):
    """Row-normalize sparse matrix"""
    rowsum = np.array(mx.sum(1))
    r_inv = np.power(rowsum, -1).flatten()
    r_inv[np.isinf(r_inv)] = 0.
    r_mat_inv = sp.diags(r_inv)
    mx = r_mat_inv.dot(mx)
    return mx

time1 = time.time()

n_samples = 1005


temp = np.zeros((n_samples, 3))

for i in range(20):
    A__ = np.multiply(np.random.binomial(1, 0.95, n_samples * n_samples).reshape(n_samples, n_samples), A)
    norm_A = normalize(A_)
    
    X = np.random.randn(n_samples, 3)
    for i in range(50):
        X = norm_A @ X
#         X = X - X.mean(axis=0)
#         X = X @ np.diag(np.array(X.std(axis=0)).flatten() ** (-1))
    temp = np.hstack([temp,X])
    
time2 = time.time()
# print(time2 - time1)

temp = temp[:, 3:]
temp = temp @ np.diag(temp.std(axis=0) ** (-1))

kmeans = KMeans(n_clusters=10).fit(temp)
print(fowlkes_mallows_score(label, kmeans.labels_))

# W = reduce(X)
# print(temp.shape, X, W)
# plt.figure(figsize=(6, 4))
# plt.scatter(1000*W[:, 0], 1000000*W[:, 1], cmap=plt.cm.Spectral)
```
