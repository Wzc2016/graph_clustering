```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import scipy.sparse as sp
from matplotlib.animation import FuncAnimation
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import SpectralClustering
from sklearn.cluster import KMeans
from sklearn import mixture
import time
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
G = nx.karate_club_graph()
np.random.seed(122)
pos = np.random.randn(34, 35)

def reduce(pos):
    model = PCA(n_components=2)
    W = model.fit_transform(pos)
    H = model.components_
    return W

pos2 = reduce(pos)
diction = {}
for i in range(pos2.shape[0]):
    diction[i] = pos2[i]
#diction
class1 = [4, 6, 5, 10, 16]
class2 = [2, 9, 24, 25, 27, 28, 31]
class3 = [0, 1, 3, 7, 11, 12, 13, 17, 19, 21]
class4 = [8, 14, 15, 18, 20, 22, 23, 26, 29, 30, 32, 33]
node_color = [0 for i in range(34)]
for i in range(34):
    if i in class1:
        node_color[i] = 'red'
    elif i in class2:
        node_color[i] = 'blue'
    elif i in class3:
        node_color[i] = 'green'
    else:
        node_color[i] = 'purple'
    
    
#pos = nx.spring_layout(G)

nx.draw(G, diction,
        node_color=node_color,
        alpha = 0.8,
        with_labels=True)

def draw2(pos):
#     model = TSNE(n_components=2, random_state=0)
#     Y = model.fit_transform(pos) 
    plt.figure(figsize=(6, 4))
    Y = reduce(pos)
    plt.scatter(Y[:, 0], Y[:, 1], c=node_color, cmap=plt.cm.Spectral)
```

```python
A = nx.adjacency_matrix(G)

time1 = time.time()
# clustering = SpectralClustering(n_clusters=2, affinity='nearest_neighbors').fit(A)
D = np.diag(np.sum(np.array(A.todense()), axis=1))
L = D - A
vals, vecs = np.linalg.eig( np.linalg.inv(D) @ L)
vecs = vecs[:,np.argsort(vals)]
vals = vals[np.argsort(vals)]

vecs = vecs[:, 1:3]
print(vals)

clustering = KMeans(n_clusters=4).fit(vecs)
time2 = time.time()
clustering.labels_

print(time2-time1)
print(adjusted_rand_score(clustering.labels_, node_color))
```

```python
draw2(vecs)
```

```python
def normalize(mx):
    """Row-normalize sparse matrix"""
    rowsum = np.array(mx.sum(1))
    r_inv = np.power(rowsum, -1).flatten()
    r_inv[np.isinf(r_inv)] = 0.
    r_mat_inv = sp.diags(r_inv)
    mx = r_mat_inv.dot(mx)
    return mx

time1 = time.time()
norm_A = normalize(np.eye(34) + A)
X = np.random.randn(34, 20)
for i in range(100):
    X = norm_A @ X * 2
    X = X - np.ones((X.shape[0], 1)) @ X.mean(axis=0)
X = reduce(X)
time2 = time.time()
print(time2-time1)

kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
print(adjusted_rand_score(kmeans.labels_, node_color))
```
